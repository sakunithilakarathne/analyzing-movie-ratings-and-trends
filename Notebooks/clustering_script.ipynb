```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import silhouette_score
```

```python
movies_data = pd.read_csv("dataset_for_clustering.csv")
movies_data.head(3)
```

```python
features = movies_data[['Domestic Gross', 'Movie Runtime', 'IMDB Rating']]
```

```python
# Elbow Method to find the optimal number of clusters
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# Apply K-Means with the chosen number of clusters
kmeans = KMeans(n_clusters=3, random_state=42)
movies_data['cluster'] = kmeans.fit_predict(features)

# Display the clustered data
print(movies_data[['IMDb ID', 'cluster']])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(movies_data['Domestic Gross'], movies_data['Movie Runtime'], c=movies_data['cluster'], cmap='viridis', s=50)
plt.xlabel('Domestic Gross')
plt.ylabel('Movie Runtime')
plt.title('Movie Clusters')
plt.colorbar(label='Cluster')
plt.show()
```
