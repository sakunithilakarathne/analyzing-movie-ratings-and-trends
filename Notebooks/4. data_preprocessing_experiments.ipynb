```python
import pandas as pd
```

```python
movies_data = pd.read_csv("complete_movie_data.csv")
#movies_data.head()
```

```python
# Cheking the data types
movies_data.dtypes
```

```python
# Checking for missing values
movies_data.isna().sum()
```

```python
# Removing min from Movie Runtime column and changing type to int
movies_data["Movie Runtime"] = movies_data["Movie Runtime"].str.split(" ").str[0].astype(int)
movies_data.head()
```

```python
# Removing $ and , from Domestic Gross and Changing type to int
movies_data["Domestic Gross"] = movies_data["Domestic Gross"].str.replace("$","")
movies_data["Domestic Gross"] = movies_data["Domestic Gross"].str.replace(",","")
movies_data["Domestic Gross"] = pd.to_numeric(movies_data["Domestic Gross"], errors='coerce').astype("Int64") 
movies_data.head()
```

```python
# Changing IMDB rating
movies_data["IMDB Rating"] = movies_data["IMDB Rating"].str.split("/").str[0]
movies_data["IMDB Rating"] = pd.to_numeric(movies_data["IMDB Rating"], errors="coerce")
movies_data.head()
```

```python
# Changing RT Rating
movies_data["RT Rating"] = movies_data["RT Rating"].str.replace("%","")
movies_data["RT Rating"] = pd.to_numeric(movies_data["RT Rating"], errors="coerce").astype("Int64")
movies_data.head()
```

```python
# Changing Metacritic rating
movies_data["Metacritic Rating"] = movies_data["Metacritic Rating"].str.split("/").str[0]
movies_data["Metacritic Rating"] = pd.to_numeric(movies_data["Metacritic Rating"], errors="coerce").astype("Int64")
movies_data.head()
```

```python
#Changing IMDb Vote count
movies_data["IMDB Vote Count"] = movies_data["IMDB Vote Count"].str.replace(",","")
movies_data["IMDB Vote Count"] = pd.to_numeric(movies_data["IMDB Vote Count"], errors='coerce').astype("Int64")
movies_data.head()
```
